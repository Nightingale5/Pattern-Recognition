# Perceptron
Perceptron is a linear classifier.  
Its output is the linear combination of the input vector through a step function.  

$\omega$: weights  
b: bias


```python
# import scipy
# import sklearn
import matplotlib.pyplot as plt
import numpy as np

max_iteration = 500 # maximum iteration of learning

num = 4 # number of samples
x = np.random.randint(-10,10,[num,2])  # points
y = np.zeros(num)
y[0:int(num/2)] = 1
y[int(num/2):] = -1 # half positive, half negative

for i in range(num):
    if(y[i]==-1):
        plt.plot(x[i][0],x[i][1],'ko')
    else:
        plt.plot(x[i][0],x[i][1],'kx')

# initialize weights
w = np.random.randn(2)
b = np.random.randn()

lr = 0.3 # learning rate
f = 0  # flag to detect whether all samples are correctly classified
 
# gradient descent
for i in range(max_iteration):
    f = 1
    for i in range(len(x)):
        if((np.dot(w,x[i]) + b) * y[i] < 0):
            f = 0
            w = w + lr * y[i] * x[i] 
            b = b + lr * y[i] # update weights
    if (f == 1):
        break;

# calculate parameters of the hyperplane and plot it
slope = -w[0]/w[1]
intercept = -b/w[1]
print('slope = ',slope)
print('intercept = ',intercept)
X = np.linspace(-10,10,300)
Y = slope * X + intercept
plt.plot(X,Y)

plt.show()
```
